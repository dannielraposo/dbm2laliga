# 0. Importing libraries

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import fpgrowth
```

# 1. Frequent Pattern Mining

```python
playersFPM = pd.read_csv(r'datacsv/playerFPM.csv')
playersFPM.head()
```

```python
apriori(playersFPM, min_support=0.35, use_colnames=True)
```

```python
fpgrowth(playersFPM, min_support=0.35, use_colnames=True)[0:30]
```
